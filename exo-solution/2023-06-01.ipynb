```python
import re
```

```python
def is_prime(n):
    for i in range(2,n):
        if n%i == 0:
            return False
    return True
```

```python
is_prime(97)
```

```python
def re_is_prime(n):
    x = "1"*n
    out = re.match(r"^(11+?)\1+$",x)
    return not bool(out)
```

```python
from random import randint

# test if the two functions give the same result
k = 100
high = 10000
for _ in range(k):
    n = randint(2,high)
    assert is_prime(n) == re_is_prime(n), f"Fail for n = {n}."

print(f"The two functions are probably equal at least between 2 and {high}.")
```

```python
def lonely(A):
    tmp = 0
    for a in A:
        tmp ^= a
    return tmp
```

```python
lonely([3,4,3,2,4,2,6,5,7,1,7,5,1])
```

```python
def Fib_rec(n: int):
    if n == 0:
        return 0
    if n == 1:
        return 1
    fib_n = Fib_rec(n - 1) + Fib_rec(n - 2)
    return fib_n
```

```python
Fib_rec(1000) # try to be patient if you can :p
```

```python
def Fib_memoized(n, memo = {}):
    if n == 0:
        return 0
    if n == 1:
        return 1

    if n in memo:
        return memo[n]

    fib_n = Fib_memoized(n-1) + Fib_memoized(n-2)
    memo[n] = fib_n
    return fib_n
```

```python
Fib_memoized(1000)
```

```python
Fib_memoized(10000)
```

```python
def Fib(n):
    dp = [None]*(n+1)
    dp[0] = 0
    dp[1] = 1
    for i in range(2,n+1):
        dp[i] = dp[i-1] + dp[i-2]
    return dp[n]
```

```python
Fib(10000)
```

```python
def Fib(n):
    if n in (0,1):
        return n
    i_2 = 0
    i_1 = 1
    for i in range(2,n+1):
        U = i_1 + i_2
        i_1, i_2 = U, i_1
    return U
```
